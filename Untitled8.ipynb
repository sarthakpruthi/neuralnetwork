```python
import torch
import torch.nn as nn
import torch.optim as optim
from fastai import *
from fastai.basics import *
import pandas as pd
```

```python
data['I']=round(data['I'],2)
data['V']=round(data['V'],2)
data['d']=round(data['d'],2)
```

```python
print(data.count());data.isna().count()
```

```python
x=data[["I","V"]]
y=data["d"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python

x_train,y_train,x_test,y_test=map(torch.tensor,(x_train.values,y_train.values,x_test.values,y_test.values))
```

```python
x_train=x_train.float()
x_test=x_test.float()
y_train=y_train.float()
y_test=y_test.float()
```

```python
n,c=x_train.shape
x_train.shape,y_train.min(),y_train.max()
```

```python
train_ds=TensorDataset(x_train,y_train)
test_ds=TensorDataset(x_test,y_test)
data=DataBunch.create(train_ds,test_ds,bs=64)
```

```python
x,y=next(iter(data.train_dl))
x.shape,y.shape
```

```python
class neural(nn.Module):
    def __init__(self):
        super().__init__()
        self.lin1=nn.Linear(2,5,bias=True)
        self.lin2=nn.Linear(5,20,bias=True)
        self.lin3=nn.Linear(20,5,bias=True)
        self.lin4=nn.Linear(5,1,bias=True)
    def forward(self,xb):
        x=self.lin1(xb)
        x=F.relu(x)
        return self.lin4(x)
```

```python
model=neural()
model
```

```python
model.lin1
```

```python
[p.shape for p in model.parameters()]
```

```python
loss_func=nn.MSELoss()
```

```python
def update(x,y,lr):
    opt=optim.Adam(model.parameters(),lr)
    y_hat=(model(x))
    loss=loss_func(y_hat,y)
    loss.backward()
    opt.step()
    opt.zero_grad()
    return loss.item()
```

```python
losses=[update(x,y,1e-3) for x,y in data.train_dl]
```

```python
losses
```

```python
def accuracy_1(input:Tensor, targs:Tensor)->Rank0Tensor:
    targs = targs.view(-1).long()
    n = targs.shape[0]
    input = input.argmax(dim=-1).view(n,-1)
    targs = targs.view(n,-1)
    return (input==targs).float().mean()
```

```python
learn=Learner(data,neural(),loss_func=loss_func,metrics=accuracy_1)
```

```python
learn.fit_one_cycle(1,1e-3)
```
